```python
from pykeen.triples import TriplesFactory
import pandas as pd
import pickle
from knowledgebase import KB
```

```python
kb = KB()
with open('kb.pickle', 'rb') as handle:
    kb = pickle.load(handle)

df = pd.DataFrame([r.values() for r in kb.relations], columns = ['head', 'type', 'tail'])
```

```python
tf = TriplesFactory.from_labeled_triples(df[['head', 'type', 'tail']][0:100].values)
```

```python
training, testing, validation = tf.split([.8, .1, .1])
```

```python
from pykeen.pipeline import pipeline

result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model = 'TransE',
    epochs = 1000,
    dimensions = 512,
    random_seed=1,
    device='cpu')

result.plot()
```

```python
from pykeen.models.predict import get_tail_prediction_df

df = get_tail_prediction_df(
    model=result.model,
    head_label="Adverity",
    relation_label="investment_size",
    triples_factory=result.training,
    add_novelties=False,
)
print(df.head(5))
```
